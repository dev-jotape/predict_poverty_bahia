```python
import pandas as pd
import numpy as np
```

```python
pib = pd.read_excel('../input/pib_2010.xlsx', skiprows=3)
pib.columns = ['city_code', 'city', 'pib']
pib[['city', 'state']] = pib['city'].str.split('(', expand=True)
pib['state'] = pib['state'].str.replace(')', '')
pib = pib.dropna()
pib = pib[pib['state'] == 'BA']
pib['pib'] = pib['pib'].astype(float)

pib.head()
```

```python
populacao = pd.read_excel('../input/population_2011.xlsx', skiprows=3)
populacao.columns = ['city_code', 'city', 'population']
populacao[['city', 'state']] = populacao['city'].str.split('(', expand=True)
populacao['state'] = populacao['state'].str.replace(')', '')
populacao = populacao.dropna()
populacao = populacao[populacao['state'] == 'BA']
populacao['population'] = populacao['population'].astype(int)

populacao
```

```python
join = pd.merge(populacao, pib, on=['city_code'])
df = join[['city_code', 'pib', 'population']]
df['pib_per_capta'] = (df['pib'] / df['population']).round(3)
df.head()
```

```python

df.to_csv('../model/pib_ba.csv', index=False)
```
