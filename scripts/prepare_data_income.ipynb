```python
import pandas as pd
import numpy as np
```

```python
income = pd.read_excel('../input/income_2010.xlsx', skiprows=6)
income.head()
```

```python
income[['city', 'state']] = income['Município'].str.split('(', expand=True)
income['state'] = income['state'].str.replace(')', '')
income.head()
```

```python
income.drop(['Classes de rendimento nominal mensal', 'Grupo de idade', 'Município'], axis=1, inplace=True)
income = income[income.state == 'BA']
income
```

```python
income.columns = ['city_code', 'income', 'city', 'state']
income.dropna(inplace = True)
income['city_code'] = income['city_code'].astype(int)
income
```

```python
income.to_csv('../model/income_avg_ba.csv', index=False)
```
