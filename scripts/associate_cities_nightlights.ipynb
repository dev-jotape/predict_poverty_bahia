```python
import pandas as pd
import geopandas as gpd
```

```python
nighttime_lights = pd.read_csv('../model/nightlights_ba.csv')
nighttime_lights
```

```python
shapefile_ba = gpd.read_file('../input/shapes/BA/29MUE250GC_SIR.shp')
shapefile_ba
```

```python
gdf = gpd.GeoDataFrame(
      nighttime_lights, geometry=gpd.points_from_xy(nighttime_lights.long, nighttime_lights.lat))
```

```python
points_in_polygons_gdf = gpd.sjoin(gdf, shapefile_ba, op='within', how='inner')
```

```python
points_in_polygons_gdf.to_csv('../model/nightlights_per_city.csv', index=False)
```
