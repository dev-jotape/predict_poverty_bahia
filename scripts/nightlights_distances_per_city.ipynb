```python
import pandas as pd
import numpy as np
```

```python
nightlights = pd.read_csv('../model/nightlights_per_city.csv')
nightlights
```

```python
city_center_coordinates = pd.read_csv('../model/city_center_coordinates.csv')
city_center_coordinates
```

```python
def euclideanDistance(lat, long, code):
    city_center = city_center_coordinates[city_center_coordinates['CD_GEOCODM'] == code].iloc[0]
    if city_center is None:
        return 0
    else:
        return np.sqrt((lat - city_center.LAT)**2 + (long - city_center.LONG)**2)
```

```python
nightlights['distance'] = nightlights.apply(lambda x: euclideanDistance(x.lat, x.long, x.CD_GEOCMU), axis=1)
nightlights
```

```python
nightlights.drop(['index_right', 'distance'], inplace=True, axis=1)
```

```python
nightlights.columns = ['lat','long','radiance','city','city_code','dist_to_city']
nightlights
```

```python
nightlights.to_csv('../model/coordinates_with_distances.csv', index=False)
```
